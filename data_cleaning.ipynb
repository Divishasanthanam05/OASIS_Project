```python
import pandas as pd
import numpy as np


df = pd.read_csv('password.csv')


df.head()


df.info()
```

```python

print(f"Duplicate rows found: {df.duplicated().sum()}")


df = df.drop_duplicates()
```

```python

print(df.isnull().sum())


df['name'] = df['name'].fillna("Unknown Listing")
df['host_name'] = df['host_name'].fillna("Unknown Host")


df['reviews_per_month'] = df['reviews_per_month'].fillna(0)

```

```python

df['room_type'] = df['room_type'].str.lower().str.strip()


df['last_review'] = pd.to_datetime(df['last_review'])

df['price'] = df['price'].astype(float)
```

```python

print(df['price'].describe())


df = df[df['price'] > 0]


upper_limit = df['price'].quantile(0.99) 
df = df[df['price'] <= upper_limit]
```

```python

df.to_csv('AB_NYC_2019_CLEANED.csv', index=False)
print("Data cleaning complete! File saved successfully.")
```

```python

import pandas as pd

df = pd.read_csv("password.csv")
print(df.head())
```

```python

df['name'] = df['name'].fillna("Unknown Listing")
df['host_name'] = df['host_name'].fillna("Unknown Host")


df['reviews_per_month'] = df['reviews_per_month'].fillna(0.0)


df['last_review'] = df['last_review'].fillna(pd.NaT)


print("--- REMAINING MISSING VALUES ---")
print(df.isnull().sum())
```

```python

df = df.drop(columns=['id', 'host_name', 'last_review'])
```

```python

df = df[df['price'] > 0]


max_acceptable_price = df['price'].quantile(0.99)
df = df[df['price'] <= max_acceptable_price]

print(f"Data cleaned! Maximum price is now limited to: ${max_acceptable_price}")
```

```python

df.to_csv('Oasis_Project3_Cleaned.csv', index=False)
print("Oasis Project 3 Data Cleaning Complete! File saved successfully.")
```
